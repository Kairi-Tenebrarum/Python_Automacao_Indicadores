### Passo 1 - Importar Arquivos e Bibliotecas

```python
# importando as bibliotecas

import pandas as pd
import pathlib
import win32com.client as win32
```

```python
# importando as bases de dados

emails = pd.read_excel(r'Bases de Dados\Emails.xlsx')
lojas = pd.read_csv(r'Bases de Dados\Lojas.csv', encoding='latin1', sep=';')
vendas = pd.read_excel(r'Bases de Dados\Vendas.xlsx')
display(emails)
display(lojas)
display(vendas)
```

### Passo 2 - Definir Criar uma Tabela para cada Loja e Definir o dia do Indicador

```python
# incluir arquivo lojas em arquivo vendas

vendas = vendas.merge(lojas, on='ID Loja')
display(vendas)
```

```python


dicionario_lojas = {}
for loja in lojas['Loja']:
    dicionario_lojas[loja] = vendas.loc[vendas['Loja'] == loja, :]

display(dicionario_lojas['Rio Mar Recife'])
display(dicionario_lojas['Shopping Vila Velha'])
```

```python
dia_indicador = vendas['Data'].max()
print(dia_indicador)
print(f'{dia_indicador.day}/{dia_indicador.month}')
```

### Passo 3 - Salvar a planilha na pasta de backup

```python
# identificando se a pasta referente a loja já existe
caminho_backup = pathlib.Path(r'Backup Arquivos Lojas')

arquivos_pasta_backup = caminho_backup.iterdir()

lista_nomes_backup = [arquivo.name for arquivo in arquivos_pasta_backup]


# testando pra ver se existe o arquivo (nome da loja)
# if (nome da loja) in lista_nomes_backup:
#     print('Existe')


for loja in dicionario_lojas:
    if loja not in lista_nomes_backup:
        nova_pasta = caminho_backup / loja
        nova_pasta.mkdir()

    # salvando dentro da pasta
    nome_arquivo  = f'{dia_indicador.month}_{dia_indicador.day}_{loja}.xlsx'
    local_arquivo = caminho_backup / loja / nome_arquivo
    dicionario_lojas[loja].to_excel(local_arquivo)
```

### Passo 6 - Automatizar todas as lojas

```python
def moeda(preco=0, moeda='R$'):
    return f'{moeda}{preco:,.2f}'

for loja in dicionario_lojas:
    vendas_loja = dicionario_lojas[loja]
    vendas_loja_dia = vendas_loja.loc[vendas_loja['Data'] == dia_indicador, :]


    # Faturamento

    # Faturamento ano

    faturamento_ano = vendas_loja['Valor Final'].sum()
    print(f'O faturamento no ano da loja {loja} foi de: {moeda(faturamento_ano)}')


    # Faturamento dia

    faturamento_dia = vendas_loja_dia['Valor Final'].sum()
    print(f'O faturamento no dia da loja {loja} foi de: {moeda(faturamento_dia)}')


    # Diversidade de produtos
    print('\n')

    # Diversidade de produtos no ano

    qtde_produtos_ano = len(vendas_loja['Produto'].unique())
    print(f'A variedade de produtos que a loja {loja} vendeu no ano foi de: {qtde_produtos_ano} unidades')


    # Diversidade de produtos no dia

    qtde_produtos_dia = len(vendas_loja_dia['Produto'].unique())
    print(f'A variedade de produtos que a loja {loja} vendeu no dia foi de: {qtde_produtos_dia} unidades')


    # Ticket médio
    print('\n')

    # Ticket médio do ano

    valor_venda = vendas_loja.groupby('Código Venda').sum() # Somando todos os valores da coluna Código Venda do ano
    ticket_medio_ano = valor_venda['Valor Final'].mean()    # Tirando a média da soma da coluna Valor Final do ano
    print(f'O ticket médio do ano da loja {loja} foi de: {moeda(ticket_medio_ano)}')


    # Ticket médio do dia

    valor_venda_dia = vendas_loja_dia.groupby('Código Venda').sum() # Somando todos os valores da coluna Código Venda do dia
    ticket_medio_dia = valor_venda_dia['Valor Final'].mean()        # Tirando a média da soma da coluna Valor Final do dia
    print(f'O ticket médio do dia da loja {loja} foi de: {moeda(ticket_medio_dia)}')
    
    
    
    outlook = win32.Dispatch('outlook.application')

    # Enviando o email para a(o) gerente da loja, nesse caso, Norte Shopping
    nome = emails.loc[emails['Loja']==loja, 'Gerente'].values[0]
    mail = outlook.CreateItem(0)
    mail.to = emails.loc[emails['Loja']==loja, 'E-mail'].values[0]
    mail.Subject = f'OnePage Dia {dia_indicador.day}/{dia_indicador.month} - Loja {loja}'
    # Usarei o mail.HTMLBody ao invés do mail.Body
    # mail.Body = 'Ainda para ser escrito, agradeço sua paciência'

    # Criando variáveis de cor
    # ------------------------------------------------------------------------
    # Cor para saber se a meta diária do faturamento foi concluída com sucesso
    if faturamento_dia >= meta_faturamento_dia:
        cor_faturamento_dia = 'green'
    else:
        cor_faturamento_dia = 'red'

    # Cor para saber se a meta anual do faturamento foi concluída com sucesso
    if faturamento_ano >= meta_faturamento_ano:
        cor_faturamento_ano = 'green'
    else:
        cor_faturamento_ano = 'red'

    # Cor para saber se a meta diária das quantidades de produtos foi concluída com sucesso
    if qtde_produtos_dia >= meta_qtde_produtos_dia:
        cor_qtde_dia = 'green'
    else:
        cor_qtde_dia = 'red'

    # Cor para saber se a meta anual das quantidades de produtos foi concluída com sucesso
    if qtde_produtos_ano >= meta_qtde_produtos_ano:
        cor_qtde_ano = 'green'
    else:
        cor_qtde_ano = 'red'

    # Cor para saber se a meta diária do ticket médio foi concluída com sucesso
    if ticket_medio_dia >= meta_ticket_medio_dia:
        cor_ticket_dia = 'green'
    else:
        cor_ticket_dia = 'red'

    # Cor para saber se a meta anual do ticket médio foi concluída com sucesso
    if ticket_medio_ano >= meta_ticket_medio_ano:
        cor_ticket_ano = 'green'
    else:
        cor_ticket_ano = 'red'
    # ---------------------------------------------------------------------------


    mail.HTMLBody = f'''
    <p> Bom dia, {nome} </p>

    <p> O resultado do dia <strong>({dia_indicador.day}/{dia_indicador.month})</strong> da <strong>loja {loja}</strong> foi: </p>

    <!-- Tabela -->
    <table>
        <tr>
            <th>Indicador</th>
            <th>Valor Dia</th>
            <th>Meta Dia</th>
            <th>Cenário Dia</th>
        </tr>
        <tr>
            <td>Faturamento</td>
            <td style="text-align: center">R${faturamento_dia}</td>
            <td style="text-align: center">R${meta_faturamento_dia}</td>
            <td style="text-align: center"><font color="{cor_faturamento_dia}">◙</font></td>
        </tr>
        <tr>
            <td>Diversidade</td>
            <td style="text-align: center">{qtde_produtos_dia}</td>
            <td style="text-align: center">{meta_qtde_produtos_dia}</td>
            <td style="text-align: center"><font color="{cor_qtde_dia}">◙</font></td>
        </tr>
        <tr>
            <td>Ticket Médio</td>
            <td style="text-align: center">R${ticket_medio_dia}</td>
            <td style="text-align: center">R${meta_ticket_medio_dia}</td>
            <td style="text-align: center"><font color="{cor_ticket_dia}">◙</font></td>
        </tr>
    </table>

    <br>

    <table>
        <tr>
            <th>Indicador</th>
            <th>Valor Ano</th>
            <th>Meta Ano</th>
            <th>Cenário Ano</th>
        </tr>
        <tr>
            <td>Faturamento</td>
            <td style="text-align: center">R${faturamento_ano}</td>
            <td style="text-align: center">R${meta_Faturamento_ano}</td>
            <td style="text-align: center"><font color="{cor_faturamento_ano}">◙</font></td>
        </tr>
        <tr>
            <td>Diversidade</td>
            <td style="text-align: center">{qtde_produtos_ano}</td>
            <td style="text-align: center">{meta_qtde_produtos_ano}</td>
            <td style="text-align: center"><font color="{cor_qtde_ano}">◙</font></td>
        </tr>
        <tr>
            <td>Ticket Médio</td>
            <td style="text-align: center">R${ticket_medio_ano}</td>
            <td style="text-align: center">R${meta_ticket_medio_ano}</td>
            <td style="text-align: center"><font color="{cor_ticket_ano}">◙</font></td>
        </tr>
    </table>

    <p> Segue em anexo a planilha com todos os dados para mais detalhes. </p>

    <p> Qualquer dúvida estou à disposição. </p>

    <p> Att., </p>
    <p> Guilherme Thomé </p>
    '''

    # Anexos
    attachment = pathlib.Path.cwd() / caminho_backup / loja / f'{dia_indicador.month}_{dia_indicador.day}_{loja}.xlsx'
    # Para o Outlook ler esse caminho é necessário transformar ele obrigatoriamente para string, pois, caso contrério, dará parâmetro inválido
    mail.Attachments.Add(str(attachment))

    mail.send()
```
